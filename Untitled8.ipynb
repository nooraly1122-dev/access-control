```python
!pip install face_recognition
import cv2
import face_recognition
import numpy as np
from google.colab import files
import os
```

```python
print("Upload AUTHORIZED personnel images (name.jpg format)")
files.upload()

authorized_encodings = []
authorized_names = []

for filename in os.listdir('.'):
    if filename.endswith(('.jpg', '.jpeg', '.png')):
        name = os.path.splitext(filename)[0]
        image = face_recognition.load_image_file(filename)
        encoding = face_recognition.face_encodings(image)
        if encoding:
            authorized_encodings.append(encoding[0])
            authorized_names.append(name)
            print(f"✅ Authorized: {name}")
```

```python
import datetime

def log_access(name, access_granted):
    with open('access_log.txt', 'a') as f:
        status = "GRANTED" if access_granted else "DENIED"
        timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        f.write(f"{timestamp} - {name} - {status}\n")
```

```python
print("\n📸 Upload face at entry point")
files.upload()

for test_img in os.listdir('.'):
    if test_img.endswith(('.jpg', '.jpeg', '.png')) and test_img not in [f for f in os.listdir('.') if f.endswith(('.jpg', '.jpeg', '.png'))][:len(authorized_names)]:
        entry_face = face_recognition.load_image_file(test_img)
        entry_encodings = face_recognition.face_encodings(entry_face)

        if entry_encodings:
            matches = face_recognition.compare_faces(authorized_encodings, entry_encodings[0])

            if True in matches:
                match_index = matches.index(True)
                name = authorized_names[match_index]
                print(f"🔓 ACCESS GRANTED - Welcome {name}")
                log_access(name, True)

                # Show result
                img_bgr = cv2.cvtColor(entry_face, cv2.COLOR_RGB2BGR)
                face_locations = face_recognition.face_locations(entry_face)
                for (top, right, bottom, left) in face_locations:
                    cv2.rectangle(img_bgr, (left, top), (right, bottom), (0,255,0), 3)
                    cv2.putText(img_bgr, f"GRANTED: {name}", (left, top-10),
                               cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0,255,0), 2)
                from google.colab.patches import cv2_imshow
                cv2_imshow(img_bgr)
            else:
                print("🔴 ACCESS DENIED - Unauthorized person")
                log_access("Unknown", False)
        else:
            print("❌ No face detected")
        break

print("\n📋 Access Log:")
with open('access_log.txt', 'r') as f:
    print(f.read())
```
